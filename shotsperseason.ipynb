```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading Data
data=pd.read_csv('C:\\Users\\shrey\\OneDrive\\Desktop\\Project_Ronaldo\\data.csv')
```

```python
# Filtering out null values
data=data[data.is_goal.isnull()==False]
data=data[data.game_season.isnull()==False]
```

```python
# extracting required data from the dataset
shotsPerSeason=data.groupby(data.game_season)['is_goal'].size()
goalsPerSeason=data.groupby(data.game_season)['is_goal'].sum()
goalsPerShot=goalsPerSeason/shotsPerSeason
```

```python
# Plotting the Curve of goals vs seasons and shots vs seasons
plt.plot(shotsPerSeason.index.values,shotsPerSeason.values,marker='o',label='shots')
plt.plot(goalsPerSeason.index.values,goalsPerSeason.values,color="red",label='goals',linewidth=2)
plt.legend()
plt.xlabel('Seasons')
plt.ylabel('Shots and goals')
plt.title('Shots and goals per season')
plt.axis([0,20,0,1600])
plt.xticks(rotation=60)
plt.show()
```

```python
# PLotting the Curve: Number of Goals per Shot vs Seasons
plt.plot(goalsPerShot.index.values,goalsPerShot.values,color="green",marker='p',linewidth=2)
plt.xlabel('season')
plt.ylabel('goals ratio')
plt.title('Goals Per Shot')
plt.xticks(rotation=60)
plt.show()
```
